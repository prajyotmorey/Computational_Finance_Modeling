### Que_A. Randomly generate the 10000 combinations of weights 𝑤1, 𝑤2 and 𝑤3 s.t. 𝑤1 +𝑤2 + 𝑤3 = 1 and 𝑤1, 𝑤2, 𝑤3 ≥ 0. Generate a scatter plot of risk vs return for all weight combinations.

```python
#Concatanating All Daily return data of TATA,ONGC,Infy
All_Daily_Return=pd.concat([TATA["Daily_Return"],ONGC["Daily_Return"],Infy["Daily_Return"]],axis=1)
```

```python
All_Daily_Return=All_Daily_Return.values
All_Daily_Return=All_Daily_Return.T
All_Daily_Return.shape
```

```python
'Mean of Daily Return for Each Stocks, i.e TATA,ONGC,Infy'
Mean_Daily_Return=np.mean(All_Daily_Return,axis=1)
print(Mean_Daily_Return)
```

### Weight Calculation
### i.e W1+W2+W3=1 and W1,W2,W3 >=0

```python
#Produces n random weights that sum to 1
def rand_weights(n):
    rand = abs(np.random.rand(n))
    return rand / sum(rand)
```

```python
'Checking Is Constrained is Followed Or Not'
np.sum(rand_weights(3)) #Sum is 0
```

```python
#Returns the mean and standard deviation of returns for a random portfolio
def random_portfolio(returns):
    p = np.asmatrix(np.mean(returns, axis=1))
    w = np.asmatrix(rand_weights(returns.shape[0]))
    C = np.asmatrix(np.cov(returns))
    
    mu = w * p.T #Return of portfolio
    sigma = (w * C * w.T) #Risk of Portfolio
    
    return mu, sigma
```

```python
n_portfolios = 10000
means, stds= np.column_stack([
    random_portfolio(All_Daily_Return) 
    for _ in range(n_portfolios)
])
```

```python
fig = plt.figure()
plt.plot(stds, means, 'o', markersize=5)
plt.xlabel('Risk')
plt.ylabel('Return')
plt.title('Return Vs. Risk For All Portfolios of TATA, ONGC,Infy')
```

# Q_1_C:-

```python
Risk=[]
Return=[]
Weights=[]
for i in range(10000):
    p = Mean_Daily_Return
    w = (rand_weights(All_Daily_Return.shape[0]))
    C = (np.cov(All_Daily_Return))
    mu = np.dot(w, p) #Return of portfolio
    s=np.dot(w,C)
    sigma =np.dot(s,w) #Risk of Portfolio
    Weights.append(w)
    Return.append(mu)
    Risk.append(sigma)
```

```python
Weights=pd.DataFrame(Weights)
Weights.columns=['TATA','ONGC','Infy']
Weights
```

```python
Return=pd.DataFrame(Return)
Return.columns=['Return']
Return
```

```python
Risk=pd.DataFrame(Risk)
Risk.columns=['Risk']
Risk
```

```python
Portfolio=pd.concat([Risk,Return,Weights],axis=1)
```
